```python
# Import packages
%pylab inline
import numpy as np
import matplotlib.pyplot as plt



figsize = (10,6)
dpi = 170
fontsize = 10
# Configure parameters
plt.rcParams.update({'font.size': fontsize, 'figure.figsize': figsize, 'figure.dpi': dpi})

# Default tick label size
plt.rcParams['text.usetex'] = False
plt.rcParams['xtick.labelsize'] = fontsize
plt.rcParams['ytick.labelsize'] = fontsize
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['xtick.major.size'] = 4
plt.rcParams['ytick.major.size'] = 4
plt.rcParams['xtick.major.width'] = 1
plt.rcParams['ytick.major.width'] = 1

plt.rcParams['xtick.top'] = True
plt.rcParams['ytick.right'] = True
plt.rcParams['axes.linewidth'] = 1
```

Function to model:
$$\phi(L_x) = \frac{d\phi(L_x)}{dlogL_x} = K_{norm}[(\frac{L_x}{L_*})^{\gamma_1} + (\frac{L_x}{L_*})^{\gamma_2}]^{-1}$$

```python
# LADE model parametrs:
logK_norm = -4.53 # +/- 0.07 / Mpc^-3
logL_star = 44.96 # +/- 0.06
gamma_1 = 0.62 # +/- 0.02
gamma_2 = 3.01 # +/- 0.11


```

Redshift evolution governed by 
$$\psi(L_X,z) = \psi(L_x,0) e(L_x,z)$$

with
$$ e(z,Lx) = (1+z)^{e_1}$$ for  $[z < z_c(L_x)]$ or
$$e(z,Lx) = e(z_c)(\frac{1+z}{1+z_c(L_x)})^{e_2}$$ for $ [z \geq z_c(L_x)] $

```python
# Redshift evolution parameters for LADE
z_c = 0.75 # +/- 0.09
d = -0.19 # +/- 0.02
```

```python
# Plot 
plt.figure(figsize = [4,4], dpi = 300)
def lade(L,z):
    return (10**logK_norm+(10**(d*(1+z))))*((L/10**(logL_star))**(gamma_1) + (L/10**(logL_star))**(gamma_2))**-1

L_x = np.logspace(41,46,1000)

# For z = 0.0 < z < 0.2:
plt.loglog(L_x,lade(L_x,0.1), color = 'black', linestyle = '--', label = 'LADE')
plt.legend(fontsize = 10)
plt.savefig('lade.png')
```
